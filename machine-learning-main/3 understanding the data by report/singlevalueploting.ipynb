```python
import pandas as pd
import seaborn as sns
```

```python
path = 'titanic.csv'
pr = pd.read_csv(path)
pr.head()
```

```python
#to create the chart (bar)
pr['Survived'].value_counts().plot(kind='bar')
```

```python
#to create a pie chart
pr['Survived'].value_counts().plot(kind='pie',autopct = '%.2f')
```

```python
sns.histplot(pr['Age'])
```

```python
#this is used for numerical data
import matplotlib.pyplot as plt
plt.hist(pr["Age"])
```

```python
sns.displot(pr["Age"])
```
