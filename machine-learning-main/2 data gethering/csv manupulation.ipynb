```python
import pandas as pd
```

```python
#to read the csv file
path = 'files/Netflix.csv'
pr = pd.read_csv(path)
```

```python
#to add title names in csv file if not exist
pr = pd.read_csv(path,names=['index','title','type','description','release_year','imdb_score'])
pr.head()
```

```python
#if file is seprated with other symbol using this we can access it
pr = pd.read_csv(path,sep=',')
pr.head()
```

```python
# to show specific colums 
pr = pd.read_csv(path,usecols= ['index','title','type','description','release_year','imdb_score'])
pr.head(10)
```

```python
#to skip the specific row in file
pr = pd.read_csv(path, skiprows=[1,2])
pr.head(10)
```

```python
#enter the number of rows u want to show
pr = pd.read_csv(path,nrows=5)
pr
```
