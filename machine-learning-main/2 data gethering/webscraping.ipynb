```python
final = pd.DataFrame({})
#final = []

for j in range(2,12):

    url = 'https://www.workindia.in/jobs-in-bhopal/?pg={}'.format(j)

    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url,headers=headers).text
    #webpage

    soup = BeautifulSoup(webpage,'lxml')
    items = soup.find_all('div',class_='JobItemV2')

    names=[]
    salary =[]
    location =[]
    fication = []

    for i in items :
        names.append(i.find('h2').text.strip())

        salary.append(i.find('div',class_='f14 f-bold').text.strip())

        location.append(i.find('div',class_='JobDetail LocationDetail text-secondary').text.strip())

        fication.append(i.find('div',class_='JobDetail QualificationDetail text-secondary').text.strip())

    data = {'names':names,'salary':salary,'location':location,'qualifiction':fication}
    pr = pd.DataFrame(data)

 
    final = pd.concat([final, pr], ignore_index=True)
```
