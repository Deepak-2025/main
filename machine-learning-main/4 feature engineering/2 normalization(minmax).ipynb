```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pr = pd.read_csv('wine.csv',usecols=['class','Alcohol','Malic acid'],nrows=100)
#pr = pd.read_csv('wine.csv')
#pr.shape
#pr.head()
```

```python
#sns.kdeplot(pr['Alcohol'])
#sns.kdeplot(pr['class'])
#sns.kdeplot(pr['Malic acid'])
```

```python
# train test split  for feature scaling
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test =  train_test_split(pr.drop('class',axis=1),pr['class'],test_size=0.3,random_state=0)

x_train.shape,x_test.shape
```

```python
#  normalization = minmaxscaling
from sklearn.preprocessing  import MinMaxScaler

scalar = MinMaxScaler()

#it learn the set ,it calculate minmax of column given in x_train
scalar.fit(x_train)

# (scalar.mean_) this code show mean value of data

#storing the scaled value in train and test 
#we train only one but transform train and test both
x_train_scaled = scalar.transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
# to convert the scaled value from array to dataframe(readable)

x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
#to see the normal and scaled values 
#we can see the scaling from this code
#np.round(x_train.describe(),1)
np.round(x_train_scaled.describe(),1)
```

```python
# showing the before and after the scaling plots
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # (rows, columns, position)s
plt.scatter(x_train['Alcohol'], x_train['Malic acid'], color='blue',label='before scale')
plt.legend()

plt.subplot(1, 2, 2) 
plt.scatter(x_train_scaled['Alcohol'], x_train_scaled['Malic acid'], color='red', label='after scale')
plt.legend()
plt.tight_layout() #for neat layout
```

```python
#before and after scaling 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(x_train['Alcohol'])

plt.subplot(1, 2, 2)
sns.kdeplot(x_train_scaled['Alcohol'])

```

```python
#before and after scaling 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(x_train['Malic acid'])

plt.subplot(1, 2, 2)
sns.kdeplot(x_train_scaled['Malic acid'])

```

```python
#before and after scaling 
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(x_train)

plt.subplot(1, 2, 2)
sns.kdeplot(x_train_scaled)

```
